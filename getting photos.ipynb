```python
import pandas as pd
pd.set_option('display.max_colwidth', 1990) 
import requests
from bs4 import BeautifulSoup
```

```python
df = pd.read_csv("lost_cleaned.csv")
```

```python
df.info()
```

```python
df['contents'] = df.urls.apply(lambda url: requests.get(url).text)
```

```python
df1 = df.sort_values(by='names', ascending=True).head(15)
```

```python
def get_photo(content):
    soup = BeautifulSoup(content)
    photo = soup.select(class_="image")
    return photo[0].get('href')
```

```python
df1['photo']=df1.contents.apply(get_photo)
```
